```python
def get_time():
    utc_now = time.time()
    utc_time = datetime.utcfromtimestamp(utc_now)
    time2 = utc_time + timedelta(hours=8)
    #nm = str(time2.month)+"_"+str(time2.day)+"_"+str(time2.hour)
    return(time2)
```

```python
#params = {"u":"6436563121","m":"4324156859858956","cu":"6436563121","containerid":"231610_standard_cardlist_1487_yanyuan_index"}
#payload = {'pageNo': '1', 'pageSize': '100'}
def create_header(pid):
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:64.0) Gecko/20100101 Firefox/64.0",
            'Referer': "https://community.kaola.com/idol/" + pid + ".html",
            "Accept": "application/json",
            "Accept-Encoding": "gzip, deflate, br",
           "Accept-Language": "en,zh-CN;q=0.9,zh;q=0.8,zh-TW;q=0.7",
          "content-type": "application/json",
           "Origin": "https://community.kaola.com",
           "Host":"activity.kaola.com",

            "x-requested-with": "XMLHttpRequest"
          }
    return(headers)

def create_header_fans(pid):
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:64.0) Gecko/20100101 Firefox/64.0",
               'Referer': "https://activity.kaola.com/idols/fansRank.html?idolid=" + pid + "IDOL040&openNewPage=true",
            "Accept": "application/json, text/plain, */*",
            "Accept-Encoding": "gzip, deflate, br",
           "Accept-Language": "en,zh-CN;q=0.9,zh;q=0.8,zh-TW;q=0.7",
          "Connection": "keep-alive",
           "Origin": "https://community.kaola.com",
           "Host":"activity.kaola.com",
          }
    return(headers)
```

```python
vote = {}
orig = pd.DataFrame()
for i in range(1,100):
    pid = "%03d" % i
    pid = "IDOL" + pid
    
    a = get_time()
    c = int(float(time.mktime(a.timetuple()))*1000)
    params1 = {'idolId':pid,'t':c}
    params_fans = {'idolId':pid}
    headers_fans = create_header_fans(pid)
    res = requests.get(url_fans,params=params_fans,headers=headers_fans)
    user = json.loads(res.text)['data']
    df = pd.DataFrame(user)
    df["idolId"] = pid
    orig = pd.concat([orig,df])
    
    headers_1 = create_header(pid)
    
    res = requests.get(url1,params=params1,headers=headers_1)
    vote[i] = json.loads(res.text)['data']
```
